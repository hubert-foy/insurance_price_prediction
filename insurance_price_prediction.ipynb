***2.2.4.2 Missing Value Patterns***

Knowing the number of missing values per column will give you a feeling for the completeness of the data and will help you to quickly uncover potential problems. In order to get a better understanding of what exactly is missing, it also helps to look for missing value patterns, that is, groups of variables where, if one of the variables is missing, the others are usually missing as well. Missing value patterns are very common when the data has been generated by joining multiple tables together. A good way to identify missing value patterns is by replacing each column with an indicator column that is equal to 1 if the value is missing and 0 if the value is not missing. Each observation basically becomes a vector of ones and zeroes and if two of these vectors are identical, it means that the corresponding observations exhibit the same missing value pattern. When you have a lot of missing data, checking for missing value patterns can reveal more insights on the process that leads to the missing data.

The groupby() function is used to split a DataFrame into groups based on one or more specified categorical variables. It's often used in conjunction with statistical methods and functions, such as mean(), median(), and count(), to analyze and summarize data on the numeric variables within the groups.

Metadata: attribute values + default values

### 2.3.1 Handling Missing Values***

'age' and 'bmi' have some null values. First diplay all records have at least one null value in either age or bmi or both

```python
# Returning a list of index values, indices of rows, in the DataFrame where either the "age" value or the "bmi" value is null.
null_inds=df[(df.age.isnull()) | (df.bmi.isnull())].index.to_list()
print("List of indexes that contain at least a missing value from age or bmi columns")
print('##'*24)
print(null_inds)

print()
# Returning a DataFrame containing only the rows from the original DataFrame where either the "age" value or the "bmi" value is null.
df_nan=df[(df.age.isnull()) | (df.bmi.isnull())]
print('Print Records of indices that contain at one or more missing values in age and bmi columns')
print('##'*24)
print(df_nan)
```

- We see that age column has 4 NaNs and bmi has 2 NaNs
- SimpleImputer from sklearn allows you to update null values based on mean, median, or mode in the respective columns. We will use median to fill null values.

***Imputing missing values in a dataset by replacing them with the median value of the column***

```python
# creating an object of the SimpleImputer class from the sklearn.impute library in Python
si=SimpleImputer()
# Initialize/Instanstiate object with the following two arguments (missing_values and strategy) 
si=SimpleImputer(missing_values = np.nan, strategy="median")

# fitting the imputer to calculate the necessary statistics (median in this case) that will 
# be used to impute missing values in the dataset. 
si.fit(df[['age', 'bmi']])
  
# Filling missing values in the 'age' and 'bmi' columns of df with median of each column
# and update the columns in the original df
df[['age', 'bmi']] = si.transform(df[['age', 'bmi']])
```

```python
#Verify whether the rows are correctly imputed
# df.iloc[null_inds,:] 
df.isnull().sum()
```

***Takeways:***
- Great!. There is no "NaN"in the data
- The above code select all rows at the index positions specified by the null_inds variable, and all columns of the DataFrame. The result is a new DataFrame containing only the rows and columns specified by the index positions.
- We see that all missing values have been replaced with a median successfully.

```python
# calculating conditional probability for all pairs of columns in a pandas DataFrame
import itertools

def calculate_conditional_probabilities(df):
  # Get a list of all column pairs
  column_pairs = list(itertools.combinations(df.columns, 2))

  # Calculate the conditional probability for each pair of columns
  for column_pair in column_pairs:
    col1, col2 = column_pair
    prob = df[col1].value_counts() / df[col2].value_counts()
    print(f"Probability of {col1} given {col2}:")
    print(prob)
    print()

calculate_conditional_probabilities(df)
```

- One-hot Encoding — Represent categorical data as binary values — 0s,1s only. I prefer to use one-hot encoding over label encoding if there aren’t many unique values in the categorical feature. 

In here, I have used pandas’ one hot encoding function (get_dummies) on Region and split it into 4 columns — location_NE, location_SE, location_NW, and location_SW. One can also use label encoding for this column, however, one hot encoding gave me a better result.

```python
#Converting categorical variables to numeric form using One hot encoding
region=pd.get_dummies(df.region, prefix='location')
df = pd.concat([df, region],axis=1)
df.drop(columns='region', inplace=True)
df.sample(3)
```

***Hyperparameter Tuning***

Let’s tweak some of the algorithm parameters such as tree depth, estimators, learning rate, etc, and check for model accuracy. Manually trying out different combinations of parameter values is very time-consuming. Scikit-learn’s GridSearchCV automates this process and calculates optimized values for these parameters. I have applied GridSearch to the above 3 algorithms. Below is the one for XGBoost:

```python
# Importing packages for hyperparameter optimization
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators': [100, 80, 60, 50, 45],  
              'max_depth': [3, 4, 5, 6],
              'learning_rate' :[0.2, 0.15]
             }
                
grid = GridSearchCV(GradientBoostingRegressor(), param_grid, refit = True, verbose = 0, n_jobs=-1) #
regr_trans = TransformedTargetRegressor(regressor=grid, transformer=QuantileTransformer(output_distribution='normal'))

# fitting the model for grid search 
grid_result=regr_trans.fit(X_train, y_train)
best_params=grid_result.regressor_.best_params_
print(best_params)

#using best params to create and fit model
best_model = GradientBoostingRegressor(max_depth=best_params["max_depth"], n_estimators=best_params["n_estimators"], learning_rate=best_params["learning_rate"])
regr_trans = TransformedTargetRegressor(regressor=best_model, transformer=QuantileTransformer(output_distribution='normal'))
regr_trans.fit(X_train, y_train)
yhat = regr_trans.predict(X_test)

#evaluate metrics
r2_score(y_test, yhat), mean_absolute_error(y_test, yhat), np.sqrt(mean_squared_error(y_test, yhat))
```
